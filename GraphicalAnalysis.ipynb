# Graphical Representation of Relation between Heart Risk an Sleep Disorder

```python
import pandas as pd
df = pd.read_csv("./Datasets/Result.csv")
df.head(10)
```

```python
import numpy as np
np.corrcoef(df["Heart Risk"], df["Sleep Disorder"])[0, 1]
```

## Contour Plot

The graph illustrates that as age increases, the risk of sleep disorder and heart risk also increases. There are `two distinct clusters` indicating two groups of individuals: those with heart risk (marked as 1) and those without (marked as 0). The cluster with heart risk is more spread out, indicating a wider variation in sleep disorders among these individuals.

The graph shows that the highest sleep disorder values occur near the middle of the plot, where age is around 50 and heart risk is 1. The lowest sleep disorder values occur at the bottom left corner of the plot, where age is around 10 and heart risk is 0. The graph also shows that there is a positive correlation between age and sleep disorder, and between heart risk and sleep disorder.

```python
import seaborn as sns
sns.jointplot(data=df, x="age", y="Sleep Disorder", hue="Heart Risk", kind="kde")
```

## Count Plot

`Countplot` is used to show the frequency of each value (0 or 1) for each variable. You can also use the `hue` parameter to group the data by another variable, such as heart risk or sleep disorder. This can help you see how the distribution of one variable varies depending on the value of another variable.

```python
sns.countplot(data=df, x="Sleep Disorder", hue="Heart Risk")
```

## Heatmap

`Heatmap` to show the correlation between `two or more boolean variables`. A heatmap is a graphical representation of data where the values are represented by colors. You can use the corr method of pandas to compute the correlation matrix of your dataframe,

```python
corr = df[["age", "Sleep Disorder", "Heart Risk"]].corr()
sns.heatmap(corr, annot=True, cmap="Blues")
```

## Pair Plot

`Pairplot` to show the pairwise relationship between `multiple boolean variable`. A pairplot is a grid of `scatter plots` that show the relationship between each pair of variables in a dataset. You can also use the `hue` parameter to color the points by another variable, such as heart risk or sleep disorder. This can help you see how the relationship between two variables changes depending on the value of another variable.

```python
sns.pairplot(df[["age", "Sleep Disorder", "Heart Risk"]], hue="Heart Risk")
```
